```
import pandas as pd
import numpy as np

data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv")

print(data.head())
```

```
print(type(data))

age_series = data['Age']
fare_series = data['Fare']
survived_series = data['Survived']

print(age_series.head())
print(fare_series.head())
print(survived_series.head())
```

```
print(data.head())
print(data.info())
print(data.describe())
```

```
print(data.isnull().sum())
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data.dropna(inplace=True)
```

```
print(data.iloc[0])  # First passenger details
print(data.iloc[0:5])  # First 5 passengers
print(data[['Age', 'Fare', 'Survived']])
```

```
data.rename(columns={'Sex': 'Gender'}, inplace=True)
print(data.head())
pivot_table = data.pivot_table(values='Survived', index='Gender', columns='Pclass')
print(pivot_table)
```

```
age_above_30 = data[data['Age'] > 30]
print(age_above_30.head())
fare_above_50 = data[data['Fare'] > 50]
female_survivors = data[(data['Gender'] == 'female') & (data['Survived'] == 1)]
print(female_survivors.head())
sorted_fare = data.sort_values(by='Fare', ascending=False)
sorted_age = data.sort_values(by='Age')
sorted_survival = data.sort_values(by='Survived', ascending=False)
```

```
survival_by_gender = data.groupby('Gender')['Survived'].mean()
print(survival_by_gender)
avg_age_class = data.groupby('Pclass')['Age'].mean()
print(avg_age_class)
survival_embarked = data.groupby('Embarked')['Survived'].sum()
print(survival_embarked)
```

```
class_info = pd.DataFrame({
    'Pclass': [1, 2, 3],
    'ClassName': ['First', 'Second', 'Third']
})
merged_data = pd.merge(data, class_info, on='Pclass')
print(merged_data.head())
joined_data = data.join(class_info.set_index('Pclass'), on='Pclass')
df1 = data.head(5)
df2 = data.tail(5)

concat_df = pd.concat([df1, df2])
```

```
avg_age_survival = data.groupby('Survived')['Age'].mean()
print(avg_age_survival)
highest_fare = data['Fare'].max()
lowest_fare = data['Fare'].min()

print("Highest Fare:", highest_fare)
print("Lowest Fare:", lowest_fare)
survival_percentage = (data['Survived'].mean()) * 100
print("Survival Percentage:", survival_percentage)
```
